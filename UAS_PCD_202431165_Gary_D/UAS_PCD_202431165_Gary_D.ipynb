```python
 #1.Import semua library yang dibutuhkan
 #202431165_Gary
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2. Membaca dan menyiapkan citra asli
# 202431165_Gary
image_path = '20260712_210234.jpg'
img = cv2.imread(image_path)

# Konversi warna dari BGR (format OpenCV) ke RGB (format Matplotlib)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
h, w = img_rgb.shape[:2]
```

```python
# 3.Rotasi Citra (Rotated)
# Memutar citra sebesar 45 derajat
#202431165_Gary
center = (w // 2, h // 2)
M_rot = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img_rgb, M_rot, (w, h))
```

```python
# 4. Mengubah Ukuran Citra (Resized)
#202431165_Gary
# Mengubah ukuran menjadi setengah dari ukuran asli
resized = cv2.resize(img_rgb, (w // 2, h // 2))
```

```python
#5. Memotong Citra (Cropped)
#202431165_Gary
# Memotong citra di area tengah. Silakan ubah angka 0.15 dan 0.85 jika area potongnya kurang pas di wajahmu.
start_row, start_col = int(h * 0.15), int(w * 0.15)
end_row, end_col = int(h * 0.85), int(w * 0.85)
cropped = img_rgb[start_row:end_row, start_col:end_col]
```

```python
# 6. Membalik Citra (Flipped)
#202431165_Gary
# Membalik citra secara horizontal
flipped = cv2.flip(img_rgb, 1)
```

```python
# 7. Pergeseran Citra (Translated)
#202431165_Gary
# Menggeser citra 100 piksel ke kanan dan 100 piksel ke bawah
M_trans = np.float32([[1, 0, 100], [0, 1, 100]])
translated = cv2.warpAffine(img_rgb, M_trans, (w, h))
```

```python
# 8.Menampilkan output sesuai format soal
#202431165_Gary
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Citra Asli
axes[0, 0].imshow(img_rgb)
axes[0, 0].set_title('Citra Asli')

# After Rotated
axes[0, 1].imshow(rotated)
axes[0, 1].set_title('After Rotated')

# After Resized
axes[0, 2].imshow(resized)
axes[0, 2].set_title('After Resized')

# After Cropped
axes[1, 0].imshow(cropped)
axes[1, 0].set_title('After Cropped')

# After Flipped
axes[1, 1].imshow(flipped)
axes[1, 1].set_title('After Flipped')

# After Translated
axes[1, 2].imshow(translated)
axes[1, 2].set_title('After Translated')

plt.tight_layout()
plt.show()
```

```python
# Cell 9: Menyimpan gambar hasil per operasi

# 1. Kembalikan format warna dari RGB ke BGR untuk proses penyimpanan OpenCV
rotated_bgr = cv2.cvtColor(rotated, cv2.COLOR_RGB2BGR)
resized_bgr = cv2.cvtColor(resized, cv2.COLOR_RGB2BGR)
cropped_bgr = cv2.cvtColor(cropped, cv2.COLOR_RGB2BGR)
flipped_bgr = cv2.cvtColor(flipped, cv2.COLOR_RGB2BGR)
translated_bgr = cv2.cvtColor(translated, cv2.COLOR_RGB2BGR)

# 2. Menyimpan masing-masing gambar ke dalam file .jpg baru
cv2.imwrite('hasil_01_rotated.jpg', rotated_bgr)
cv2.imwrite('hasil_02_resized.jpg', resized_bgr)
cv2.imwrite('hasil_03_cropped.jpg', cropped_bgr)
cv2.imwrite('hasil_04_flipped.jpg', flipped_bgr)
cv2.imwrite('hasil_05_translated.jpg', translated_bgr)

print("Berhasil! Semua gambar sudah tersimpan di direktori Colab.")
```
